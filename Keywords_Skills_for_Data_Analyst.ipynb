# Part 1: Web Scraping

```python
# Import libraries 
from bs4 import BeautifulSoup
import requests
```

```python
# Create a list of websites
websites = ("https://www.northeastern.edu/graduate/blog/data-analyst-skills/",
            "https://www.indeed.com/career-advice/resumes-cover-letters/skills-for-data-analyst",
            "https://www.ziprecruiter.com/Career/Business-Data-Analyst/Resume-Keywords-and-Skills",
            "https://www.ziprecruiter.com/Career/Data-Analyst/Resume-Keywords-and-Skills",
            "https://www.jigsawacademy.com/blogs/business-analytics/skills-required-for-data-analyst/", 
            "https://www.umassglobal.edu/news-and-events/blog/data-analytics-skills",
            "https://www.finereport.com/en/data-analysis/6-key-skills-that-data-analysts-need-to-master.html",
            "https://bootcamp.cvn.columbia.edu/blog/data-analyst-skills/",
            "https://www.coursera.org/articles/in-demand-data-analyst-skills-to-get-hired")

# Create a list of tags
tags = ("li", "h3", "td", "td", "h2", "strong", "h2", "h3", "h2")

for i,j in zip(websites,tags): # Pair the websites and tags list together as shown below
    print (i,j)
```

```python
# Scrape the first 9 links

WordList = list() # Create a list to store the keywords/skills

for i,j in zip(websites,tags):
    url = i
    r = requests.get(url) # Get URL
    soup = BeautifulSoup(r.text, "html.parser") # Parser the URL with BeautifulSoup
    
    tags = soup(j)
    
    # Use for loop to pull each word and then append it to the list
    for a in tags:
        word = a.get_text() # Get the text from the tag
        WordList.append(word) # Add the keyword to WordList

print(WordList) # Show the result of the loop
```

```python
# scrape link #10

url = "https://resumeworded.com/skills-and-keywords/data-analyst-skills"
r = requests.get(url) # Get URL
soup = BeautifulSoup(r.text, "html.parser") # Parse the URl with BeautifulSoup

tags = soup(class_="skill-element")

# Use for loop to pull each word and then append it to the list
for tag in tags:
    word = tag.get_text()
    WordList.append(word) # Add the keyword to WordList

print(WordList)
```

```python
# scrape link #11

url = "https://www.indeed.com/career-advice/resumes-cover-letters/skills-for-data-analyst"
r = requests.get(url) # Get URL
soup = BeautifulSoup(r.text, "html.parser") # Parse the URl with BeautifulSoup

tags = soup(class_="styles-module--listItemText--2JjD8")

# Use for loop to pull each word and then append it to the list
for tag in tags:
    word = tag.get_text()
    WordList.append(word) # Add the keyword to WordList

print(WordList)
```

```python
print("Total items in the list: ",len(WordList))
```

# Part 2: numpy + pandas

```python
# Import libraries
import pandas as pd
import numpy as np

pd.set_option("display.max_rows", None) # Display all the rows
```

```python
# Create a dataframe and give the column name "Keyword"
df = pd.DataFrame(WordList, columns=["Keyword"])
print(df)
```

```python
""" 
2 problems:
1) Numbers, decimals, and percentages
2) Irrelevant information
""" 
```

```python
# Clean WordList
# Step 1 - Get rid of numbers, decimals, and percentages 

df["Keyword"] = df["Keyword"].map(lambda x: x.lstrip("0123456789.%")) # Remove specific set of characters
print(df)
```

```python
# Step 2 - Drop the rows that contain irrelevant information

rows = (25, 26, 27, 28, 29, 30, 31, 32, 33, 134, 135, 136, 157, 158, 159, 160, 161, 162, 173, 185, 186, 187, 188, 189, 190, 191,
        192, 193, 194, 202, 203, 204, 205, 206, 207) #Rows to be dropped

DropRows = list(rows) # Turn it into a list

df.drop(DropRows, inplace = True) # Drop rows according to the list
print(df)
```

# Part 3: Data Combine, Merge, Load

```python
df.value_counts() # Check the frequency for each keyword
```

```python
"""
4 problems:
1) Empty strings
2) Extra space in front strings
3) Inconsistency keywords/skills
4) Upper/lower case
"""
```

```python
# 1) Remove the rows that contain empty value

df["Keyword"] = df["Keyword"].replace(["", " "], [np.nan, np.nan]) # Replace enpty strings with "NaN"
df = df[df["Keyword"].notna()] # Filter out NaN values

df.value_counts() # Check the frequency for each keyword after step 1
```

```python
# 2) Get rid of empty spaces in front of the strings

df["Keyword"] = df["Keyword"].str.strip() # Remove trailing spaces from the right of the string

print(df.value_counts()) # Check the frequency for each keyword after step 2
```

```python
# 3) Replace Inconsistent keywords/skills

# Replace the value in to_replace to the value in value respectively
df.replace(to_replace= ["Structured Query Language (SQL)", "R (Programming Language)", "Python (Programming Language)",
                        "SAS Programming", "Data Visualization Tools", "Business Intelligence (BI)",
                        "Statistical programming languages", "Analytics", "Databases", "Data Visualization Skills",
                        "Statistical Programming Languages", "Data Warehouse", "Data Mining Skills", 
                        "Business Intelligence Tools", "Programming Skills", "SQL Server", "Public speaking", "Problem-solving"],
           value = ["SQL", "R", "Python", "SAS", "Data Visualization", "Business Intelligence", "Statistical Programming",
                    "Data Analytics", "Database", "Data Visualization", "Statistical Programming", "Data Warehousing", 
                    "Data Mining", "Business Intelligence", "Programming Language", "SQL", "Communication", "Problem Solving"], inplace = True)

print(df.value_counts()) # Check the frequency for each keyword after step 3
```

```python
# 4) Inconsistent upper/lower case

df["Keyword"] = df["Keyword"].str.title() # Capitalize the first letter of each word

print(df.value_counts()) # Check the frequency for each keyword after step 4
```

```python
# Top 15 list of keywords for data analyst
Top15 = df.value_counts()[0:15]
print(Top15)
```

```python
# Rest index numbering
df.reset_index(drop=True)
```

```python
print("Total items in the list after cleaning:", len(df))
```

# Part 4: Data Aggregation and Data Visualization

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Horizontal bar graph for the top 15 keywords
Top15.sort_values().plot(kind = "barh", figsize=(15, 10), color = sns.color_palette("Paired"))

TitleFont = {"family": "Times New Roman", "color": "grey", "size": 20}
xLabelFont = {"family": "Times New Roman", "color": "darkblue", "size": 16}
yLabelFont = {"family": "Times New Roman", "color": "darkblue", "size": 16}

plt.title("Most Frequently Used Keywords/Skills For Data Analyst", fontdict = TitleFont) # Graph title
plt.xlabel("Frequency", fontdict = xLabelFont) # x-asix label
plt.ylabel("Keyword", fontdict = yLabelFont) # y-axis label
plt.show() # Show grpah
```

```python
# Pie chart - compare Python and R
df.value_counts()[["Python", "R"]].plot(kind = "pie", labels = ["Python", "R"], ylabel = '', startangle = 90)
plt.show() #Show graph
```
